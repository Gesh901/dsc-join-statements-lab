## Select the names of all employees in Boston 

Hint: join the employees and offices tables. Select the first and last name.

```python
# Your code here

query ="""

SELECT firstName, lastName
  FROM employees
  JOIN offices
 USING (officeCode)
 WHERE city = "Boston";


"""

pd.read_sql(query,conn)
```

## Are there any offices that have zero employees?
Hint: Combine the employees and offices tables and use a group by. Select the office code, city, and number of employees.

```python
# Your code here

query ="""

SELECT officeCode,city, COUNT(employeeNumber) AS no_of_employees
  FROM employees
  JOIN offices
 USING (officeCode)
 GROUP BY officeCode;

"""

pd.read_sql(query,conn)
```

## Write 3 questions of your own and answer them

```python
# Answers will vary

# Example question: 
"""
How many customers are there per office?
"""

query ="""

SELECT COUNT(DISTINCT customerNumber) AS no_of_customers
  FROM customers
  JOIN employees
  ON customers.SalesRepEmployeeNumber = employees.employeeNumber
  GROUP BY officeCode;

"""

pd.read_sql(query,conn)
```

```python
"""
How many orders have been made per customer?
"""

# Your code here

query ="""

SELECT customerName, COUNT(DISTINCT orderNumber) AS no_of_orders
  FROM orders
  JOIN customers
 USING (customerNumber)
  GROUP BY customerNumber;

"""

pd.read_sql(query,conn)

```

```python
"""
What are the payment details for orders made by the customers?
"""

# Your code here

query ="""

SELECT * 
  FROM customers
  JOIN payments
 USING (customerNumber);

"""

pd.read_sql(query,conn)
```

```python
"""
How may products are in each product line 
"""

# Your code here

query ="""

SELECT productLine, COUNT(*) AS no_of_products
  FROM productlines
  JOIN products
 USING (productLine)
 GROUP BY productLine;

"""

pd.read_sql(query,conn)
```

## Level Up 1: Display the names of every individual product that each employee has sold

Hint: You will need to use multiple `JOIN` clauses to connect all the way from employee names to product names.

```python
# Your code here

query ="""

SELECT employees.employeeNumber, employees.firstName, employees.lastName, products.productName
  FROM products
       JOIN orderdetails
       ON products.productCode = orderdetails.productCode
          JOIN orders
          ON orderdetails.orderNumber = orders.orderNumber
             JOIN customers
             ON orders.customerNumber = customers.customerNumber
                JOIN employees
                ON customers.SalesRepEmployeeNumber = employees.employeeNumber;

"""

pd.read_sql(query,conn)
```

## Level Up 2: Display the number of products each employee has sold

Alphabetize the results by employee last name.

Hint: Use the `quantityOrdered` column from `orderDetails`. Also, think about how to group the data when some employees might have the same first or last name.

```python
# Your code here

query ="""

SELECT employees.employeeNumber, employees.firstName, employees.lastName, SUM(orderdetails.quantityOrdered) AS no_of_products
  FROM products
       JOIN orderdetails
       ON products.productCode = orderdetails.productCode
          JOIN orders
          ON orderdetails.orderNumber = orders.orderNumber
             JOIN customers
             ON orders.customerNumber = customers.customerNumber
                JOIN employees
                ON customers.SalesRepEmployeeNumber = employees.employeeNumber
                
  GROUP BY employees.employeeNumber
  ORDER BY lastName;              

"""

pd.read_sql(query,conn)
```

## Level Up 3: Display the names employees who have sold more than 200 different products

Hint: this is different from the previous question because the quantity sold doesn't matter, only the number of different products

```python
# Your code here

query ="""

SELECT employees.employeeNumber, employees.firstName, employees.lastName, COUNT(products.productName) AS no_of_products
  FROM products
       JOIN orderdetails
       ON products.productCode = orderdetails.productCode
          JOIN orders
          ON orderdetails.orderNumber = orders.orderNumber
             JOIN customers
             ON orders.customerNumber = customers.customerNumber
                JOIN employees
                ON customers.SalesRepEmployeeNumber = employees.employeeNumber
                
  GROUP BY employees.employeeNumber
  HAVING no_of_products > 200;              

"""

pd.read_sql(query,conn)
```
